```python
# while 1循环，一直循环
while 1:
    def s_sjx():
    # 正实心三角形
        for i in range(1,6):
            # 更改打印起始位置
            for k in range(1,6-i):
                print("",end=" ")
            # 控制列，规则是一行一列，第二行是二列
            for j in range(6-i,6):
                print("*",end=" ")
            print()


    def k_sjx():
    # 正空心三角
        for i in range(1,6):
            # 更改打印起始位置
            for k in range(1,6-i):
                print("",end=" ") 
            # 控制列，规则是第一行一列，第二行是二列
            for j in range(6-i,6):
                if i==1 or i==5 or j==6-i or j==5:
                    print("*",end=" ")
                else:
                    print(" ",end=" ")
            print()

    def s_lx():
    # 正菱形实心
        for i in range(1,5):
            # 更改打印起始位置
            for k in range(1,7-i):
                print("",end=" ")
            # 控制列，规则式第一行一列，第二行是二列
            for j in range(6-i,6):
                print("*",end=" ")
            print()

        for i in range(1,6):
            # 更改打印起始位置
            for k in range(i):
                print("",end=" ")
            for j in range(6-i):
            # 控制列，规则是第一行一列，第二行是二列
                print("*",end=" ")
            print()




    def k_lx():
    # 正菱形空心
        for i in range(1,5):
            # 更改打印起始位置
            for k in range(1,7-i):
                print("",end=" ")
            # 控制列，规则式第一行一列，第二行是二列
            for j in range(6-i,6):
                if i==1 or i==5 or j==6-i or j==5:
                    print("*",end=" ")
                else:
                    print(" ",end=" ")
            print()

        for i in range(1,6):
            # 更改打印起始位置
            for k in range(i):
                print("",end=" ")
            for j in range(6-i):
                if i==5 or j==0 or j==6-i-1:
            # 控制列，规则是第一行一列，第二行是二列
                    print("*",end=" ")
                else:
                    print(" ",end=" ")
            print()


    def s_tx():
    # 正梯形实心
        for i in range(1,6):
            for k in range(1,6-i):
                print(end=" ")
            for j in range(6-i,9):
                print("*",end=" ")
            print()

    def k_tx():
    # 正梯形空心
        for i in range(1,6):
            for k in range(1,6-i):
                print(end=" ")
            for j in range(6-i,9):
                if i==1 or i==5 or j==6-i or j==8:
                    print("*",end=" ")
                else:
                    print(" ",end=" ")            
            print()
    
    print('输入-1结束')
    shape = input('请输入正实心三角形或正空心三角形或正菱形实心或正菱形空心或正梯形实心或正梯形空心:')
    if shape =='-1':
        break
    elif shape =="正实心三角形":
        s_sjx()
    elif shape =="正空心三角形":
        k_sjx()
    elif shape =="正菱形实心":
        s_lx()
    elif shape =="正菱形空心":
        k_lx()
    elif shape =="正梯形实心":
        s_tx()
    elif shape =="正梯形空心":
        k_tx()

    else:
        print('你是不是傻！看不懂还是咋滴！')
    
```
